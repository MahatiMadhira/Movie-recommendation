Importing libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Dataset retrieval and pre-processing

```python
movie_data = pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

Feature selection

```python
selected_features = ['genres', 'keywords', 'original_title', 'tagline', 'cast', 'director']
for feature in selected_features:
  movie_data[feature] = movie_data[feature].fillna('')
```

Combine selected features

```python
combined_features = movie_data['genres'] + ' ' + movie_data['keywords'] + ' ' + movie_data['original_title'] + ' ' + movie_data['tagline'] + ' ' + movie_data['cast'] + ' ' + movie_data['director']
```

```python
print(combined_features)
```

Convert textual data into Feature Vectors using TfidVectorizer

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Using Cosine Similarity of sklearn to get similarity scores for all the movies in the dataset in order to create a recommendation system

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

Taking the user's favorite movie as input

```python
user_movie = input('What is your favorite movie? ')
```

Check is the movie entered by user exists in our dataset

```python
movie_list = movie_data['title'].tolist()
print(movie_list)
```

```python
movie_matches = difflib.get_close_matches(user_movie, movie_list)
print(movie_matches)
```

```python
movie_match = movie_matches[0]
print(movie_match)
```

Finding the index of the moveie so we can find similar movies

```python
movie_index = movie_data[movie_data.title == movie_match]['index'].values[0]
print(movie_index)
```

```python
similarity_score = list(enumerate(similarity[movie_index]))
print(similarity_score)
```

Sort the list of movies based on similarity score

```python
sorted_movies = sorted(similarity_score, key= lambda x:x[1], reverse = True)
print(sorted_movies)
```

Extract top 20 movies which are similar to user's given movie

```python
print('Suggested Movies to watch: ')
i=1

for movie in sorted_movies[1:]:
  index = movie[0]
  movie_title = movie_data[movie_data.index == index]['title'].values[0]
  if (i<21):
    print(i, '.', movie_title)
    i+=1
```

Final Movie Recommendation System

```python
user_movie = input('What is your favorite movie? ')

movie_list = movie_data['title'].tolist()

movie_matches = difflib.get_close_matches(user_movie, movie_list)

movie_match = movie_matches[0]

movie_index = movie_data[movie_data.title == movie_match]['index'].values[0]

similarity_score = list(enumerate(similarity[movie_index]))

sorted_movies = sorted(similarity_score, key= lambda x:x[1], reverse = True)

print('Suggested Movies to watch: ')
i=1

for movie in sorted_movies[1:]:
  index = movie[0]
  movie_title = movie_data[movie_data.index == index]['title'].values[0]
  if (i<21):
    print(i, '.', movie_title)
    i+=1
```
